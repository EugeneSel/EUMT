```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
dict_words = {
    "процес": 2,
    # "дисципліна": 2,
    # "обсяг\nданих": 3,
    "знання": 4,
    "наука": 1,
    # "обробка": 3,
    "аналіз": 6,
    "статистика": 7,
    "передбачення\nпрогнозування": 3,
    "алгоритм": 4,
    "галузь;\nпредметна\nобласть": 7,
    # "модель": 4,
    "міждисцип-\nлінарна;\nбагатогранна": 2,
    "професія": 3,
    # "бізнес": 1
}

dict_words = {key: value for key, value in sorted(sorted(dict_words.items(), key=lambda x: x[0]), key=lambda x: x[1], reverse=True)}
df_words = pd.DataFrame(dict_words, columns=pd.Index(dict_words.keys(), name="Term"), index=["Occur"]).T.reset_index()
```

```python
plt.figure(figsize=(24, 7))
barplot = sns.barplot(x="Term", y="Occur", data=df_words)
for p in barplot.patches:
    barplot.annotate(format(p.get_height(), '.0f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points',
                   fontsize=16)
plt.xticks(fontsize=20)
plt.yticks(fontsize=15)
plt.xlabel("Термін", fontsize=20)
plt.ylabel("Кількість джерел, в яких згадується", fontsize=20)
plt.title("Частота використання термінів в означеннях Data Science", fontsize=30)
plt.savefig("TermsFrequencies.png", dpi=300)
plt.show()
```
