```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
dict_words = {
    "process": 2,
    # "дисципліна": 2,
    # "обсяг\nданих": 3,
    "knowledge": 4,
    "science": 1,
    # "обробка": 3,
    "analysis": 6,
    "statistics": 7,
    "forecast;\nprediction": 3,
    "algorithm": 4,
    "domain\nexpertise/\nknowledge": 7,
    # "модель": 4,
    "inter-\ndisciplinary": 2,
    "profession": 3,
    # "бізнес": 1
}

dict_words = {key: value for key, value in sorted(sorted(dict_words.items(), key=lambda x: x[0]), key=lambda x: x[1], reverse=True)}
df_words = pd.DataFrame(dict_words, columns=pd.Index(dict_words.keys(), name="Term"), index=["Occur"]).T.reset_index()
```

```python
plt.figure(figsize=(24, 7))
barplot = sns.barplot(x="Term", y="Occur", data=df_words)
for p in barplot.patches:
    barplot.annotate(format(p.get_height(), '.0f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points',
                   fontsize=16)
plt.xticks(fontsize=20)
plt.yticks(fontsize=15)
plt.xlabel("Term", fontsize=20)
plt.ylabel("Number of sources, where it is mentioned", fontsize=20)
plt.title("Term use frequency in Data Science definitions", fontsize=30)
plt.savefig("TermsFrequenciesEn.png", dpi=600)
plt.show()
```
